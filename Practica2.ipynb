```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Impulso unitario
def impseq(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n-n0) == 0
    plt.stem(n, x)
    plt.title(f'Impulso u(n - {n0})')
    plt.xlabel('n')
    plt.ylabel('x(n)')
    plt.xticks(n) 
    plt.show()
    return [x, n]
```

```python
x, n = impseq(1, 0, 4)
```

# Funcion para generar el escalón unitario


```python
# Escalon
def esc(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n - n0) >= 0
    plt.stem(n, x)
    plt.title(f'Escalón Unitario u(n+{n0})')
    plt.xlabel('n')
    plt.ylabel('x(n)')
    plt.xticks(n) 
    plt.show()
    return [x, n]
```

```python
x, n = esc(1, -2, 8)
```

# Función para generar la rampa


```python
# Rampa
def ramp(n0, n1, n2, m):
    n = np.arange(n1, n2+1)
    x = m*(n-n0) * (n>=n0)
    plt.stem(n, x)
    plt.title(f'Función rampa con pendiente 1')
    plt.xlabel('n')
    plt.ylabel('x(n)')
    plt.xticks(n) 
    plt.show()
    return[x,n]
```

```python
x, n = ramp(1,-2,8,1)
```

```python
n1 = -10
n2 = 10
n = np.arange(n1, n2+1)
x4 = (
    2 * ramp(-3, n1, n2,1)[0]  # 2r(n+3)
    + -1* ramp(2, n1, n2,1)[0]  # -r(n-2)
    + -5*esc(3,n1,n2)[0]
)
plt.stem(n, x4)
plt.title(r'$x_4(n) = 2r(n+3) - r(n-2) - 5u(n-3)$')
plt.xlabel('n')
plt.ylabel(r'$x_4(n)$')
plt.xticks(n) 
plt.grid(True)
plt.show()   
```

# Punto 4, generar la secuencia 

```python
n1 = 0
n2= 17
n = np.arange(n1,n2+1)
# zeros_like :Devuelve una matriz de ceros 
# con la misma forma y tipo que una matriz dada.
X = np.zeros_like(n)
X[(n >= 0) & (n <= 5)] = n[(n >= 0) & (n <= 5)]
X[(n >= 6) & (n <= 11)] = n[(n >= 6) & (n <= 11)] - 5
X[(n >= 12) & (n <= 17)] = n[(n >= 12) & (n <= 17)] - 10

# Calcular la derivada
x5= np.diff(X, prepend=0)

# Graficar la secuencia original y su derivada discreta
plt.figure(figsize=(12, 6))

#Secuencia original
plt.subplot(2, 1, 1)
plt.stem(n, X, use_line_collection=True)
plt.title('Secuencia original $X[n]$')
plt.ylabel('X[n]')
plt.xticks(n) 
plt.grid(True)

#Derivada  de la secuencia
plt.subplot(2, 1, 2)
plt.stem(n, x5, use_line_collection=True)
plt.title('Derivada $X\'[n]$')
plt.xlabel('n')
plt.xticks(n) 
plt.ylabel('X\'[n]')
plt.grid(True)
plt.show()
```

# OBSERVACIONES.
#### En los intervalos donde X[n] crece linealmente, la derivada discreta X'[n] es aproximadamente constante e igual a la pendiente de la rampa. Sin embargo, en las transiciones entre segmentos, donde hay un salto en el valor de X[n], la derivada discreta presenta algunos picos debido a la diferencia finita entre valores consecutivos, estos picos no se observan en la derivada continua de una rampa, que sería una función constante.

### a. x5n=2xn-4+x(n)

```python
def x5n(xn):
    x5 = []
    for n in range(len(xn)):
        if n >= 4:
            x5.append(2 * xn[n - 4] + xn[n])
        else:
            x5.append(xn[n])  # Para n < 4, usamos el valor de xn[n] directamente
    return x5

x5 = x5n(xn)

# Graficar la secuencia x5n
plt.figure(figsize=(10, 6))
plt.stem(x5, use_line_collection=True)
plt.title('Secuencia x5n')
plt.xlabel('n')
plt.ylabel('x5n')
plt.grid()
plt.show()
```

#### b. x6n=0.001e^0.5n (0. 05πn) x(n+2), -20≤n≤20

```python
def x6n(xn):
    n_values = np.arange(-20, 21)
    x6 = []
    for n in n_values:
        xn_index = n + 2
        if 0 <= xn_index < len(xn):
            x6_value = 0.001 * np.exp(0.5 * n) * np.cos(0.05 * np.pi * n) + xn[xn_index]
        else:
            x6_value = 0.001 * np.exp(0.5 * n) * np.cos(0.05 * np.pi * n)
        x6.append(x6_value)
    return n_values, x6

n_values, x6 = x6n(xn)

# Gráfico de la secuencia x6n
plt.figure(figsize=(10, 6))
plt.stem(n_values, x6, use_line_collection=True)
plt.title('Secuencia x6n')
plt.xlabel('n')
plt.ylabel('x6n')
plt.grid()
plt.show()
```
