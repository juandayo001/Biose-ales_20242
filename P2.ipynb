```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Impulso unitario
def impseq(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n-n0) == 0
    return[x,n]
```

```python
x, n = impseq(-1, -3, 4)
plt.stem(n,x)
```

```python
# Escalon
def esc(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n >= n0)
    return[x,n]
```

```python
x, n = esc(0, -3, 4)
plt.stem(n,x)
```

```python
# Rampa
def ramp(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n-n0) * (n>=n0)
    return[x,n]
```

```python
x, n = ramp(0, -3, 4)
plt.stem(n,x)
```

```python
# Rampa con pendiente "m"

def ramp1(n0, n1, n2, m):
    n = np.arange(n1, n2+1)
    x = m*(n-n0) * (n>=n0)
    return[x,n]
```

```python
x, n = ramp1(0, -3, 4, 2)
plt.stem(n,x)
```
