```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.io as sio
from scipy import stats
from scipy.fftpack import fft, ifft
import os
```

# Obtenci√≥n de los datos control y de los datos de parkinson




```python
ruta_base = os.getcwd()
archivos_control = os.listdir(ruta_base +'/control')
datos_control = []
for i in range(0,len(archivos_control)):
    sujeto_i = ruta_base + '/control/' + archivos_control[i]
    datos = sio.loadmat(sujeto_i)
    datos_control.append(datos['data'])
    
archivos_parkinson = os.listdir(ruta_base + '/parkinson')
datos_parkinson = []
for i in range(0,len(archivos_parkinson)):
    sujeto_i = ruta_base + '/parkinson/' + archivos_parkinson[i]
    datos = sio.loadmat(sujeto_i)
    datos_parkinson.append(datos['data'])
    
numero_de_sujetos_c = len(archivos_control)
numero_de_sujetos_p = len(archivos_parkinson)
print(f"El grupo de control tiene {numero_de_sujetos_c} sujetos")
print(f"El grupo de parkinson tiene {numero_de_sujetos_p} sujetos")
```

# Funci√≥n para calcular la Energ√≠a primedio de cada canal.

```python
def calcular_en_prom(senal):
    senal_reshape = np.reshape(senal,(senal.shape[0],senal.shape[1]*senal.shape[2]),order='F') #convertir a 2d
    abs_squared=np.apply_along_axis(np.square,1,np.abs(senal_reshape)) # En la clase vimos que la energia era la sumatoria de todos los datos al cuadrado dentro de valor absoluto
    energia_prom =np.apply_along_axis(np.mean,1,abs_squared)
    return energia_prom 
```

### 2. Calcule la energ√≠a de cada canal promediada por √©pocas para cada sujeto, esto para ambos grupos poblacionales. Guarde esta informaci√≥n en un DataFrame de columnas 'canal' y filas '#sujeto' con los valores de energ√≠a calculados. Un DataFrame para cada grupo poblacional.

```python
data={'sujeto':[],'canal 1':[],'canal 2':[],'canal 3':[],'canal 4':[],'canal 5':[],'canal 6':[],'canal 7':[],'canal 8':[]}
for sujeto in range(len(datos_control)):
    energia = calcular_en_prom(datos_control[sujeto])
    data['sujeto'].append(sujeto+1)
    for canal in range(len(energia)): 
        data[f'canal {canal+1}'].append(energia[canal])
data_frame_control = pd.DataFrame(data)
print(data_frame_control)
```

```python
data={'sujeto':[],'canal 1':[],'canal 2':[],'canal 3':[],'canal 4':[],'canal 5':[],'canal 6':[],'canal 7':[],'canal 8':[]}
for sujeto in range(len(datos_parkinson)):
    energia = calcular_en_prom(datos_parkinson[sujeto])
    data['sujeto'].append(sujeto+1)
    for canal in range(len(energia)): 
        data[f'canal {canal+1}'].append(energia[canal])
data_frame_parkinson = pd.DataFrame(data)
print(data_frame_parkinson)
```

# 3. Prueba t

#### Necesito primero comporbar normalidad y homocedastacidad 

```python
for canal in range(1, 9): 
    print(f"Canal {canal}")
    # Energ√≠a para el grupo control y Parkinson
    energia_control = data_frame_control[f'canal {canal}']
    energia_parkinson = data_frame_parkinson[f'canal {canal}']
    
    # Prueba de Shapiro-Wilk para normalidad
    stat_control, p_control = stats.shapiro(energia_control)
    stat_parkinson, p_parkinson = stats.shapiro(energia_parkinson)
    
    print(f"Control: Estad√≠stico={stat_control}, p-valor={p_control}")
    print(f"Parkinson: Estad√≠stico={stat_parkinson}, p-valor={p_parkinson}")
    
    if p_control > 0.05:
        print("La energ√≠a del grupo control sigue una distribuci√≥n normal.")
    else:
        print("La energ√≠a del grupo control NO sigue una distribuci√≥n normal.")
        
    if p_parkinson > 0.05:
        print("La energ√≠a del grupo Parkinson sigue una distribuci√≥n normal.")
    else:
        print("La energ√≠a del grupo Parkinson NO sigue una distribuci√≥n normal.")
```

```python
# Verificar la homocedasticidad (varianzas iguales) canal por canal
for canal in range(1, 9): 
    print(f"Canal {canal}")
    # Energ√≠a para el grupo control y Parkinson
    energia_control = data_frame_control[f'canal {canal}']
    energia_parkinson = data_frame_parkinson[f'canal {canal}']
    
    # Prueba de Levene para homocedasticidad
    stat, p_levene = stats.levene(energia_control, energia_parkinson)
    print(f"Estad√≠stico Levene={stat}, p-valor={p_levene}")
    
    if p_levene > 0.05:
        print("Varianzas iguales (homocedasticidad).")
    else:
        print("Varianzas **no** iguales (heterocedasticidad).")
```

####  Dado que los grupos no siguen una distribuci√≥n normal, la prueba t no es adecuada. En su lugar, la prueba U de Mann-Whitney es m√°s apropiada y permitir√° determinar si existen diferencias significativas en los canales entre los grupos de control y Parkinson.

```python
from scipy.stats import mannwhitneyu

# Realizar la prueba U de Mann-Whitney para cada canal
for canal in range(1, 9):
    print(f"Canal {canal}")
    
    # Energ√≠a para el grupo control y Parkinson
    energia_control = data_frame_control[f'canal {canal}']
    energia_parkinson = data_frame_parkinson[f'canal {canal}']
    
    # Realizamos la prueba U de Mann-Whitney
    stat, p_value = mannwhitneyu(energia_control, energia_parkinson)
    
    print(f"Estad√≠stico U: {stat}, p-valor: {p_value}")
    
    # Interpretaci√≥n del p-valor
    if p_value < 0.05:
        print(f"Existe una diferencia significativa en el Canal {canal} entre los grupos de Control y Parkinson (p < 0.05).")
    else:
        print(f"No existe una diferencia significativa en el Canal {canal} entre los grupos de Control y Parkinson (p >= 0.05).")
```

# An√°lisis.
#### El an√°lisis que se realiz√≥ con la prueba U de Mann-Whitney en la comparaci√≥n de la energ√≠a promedio de los canales EEG mediante los grupos Control y Parkinson no mostr√≥ evidencias estad√≠sticamente significativas en los ocho canales analizado. Al observar los p-valores obtenidos todos fueron superiores a 0.05; los canales 7 y 8 fueron los que mostraron los p- valores m√°s bajos, siendo 0.0591 y 0.0752, pero a√∫n as√≠ no lograron demostrar una diferencia tan concluyente entre los grupos. Para este caso la energ√≠a de las se√±ales EEG no es un rasgo diferenciador en los sujetos con Parkinson y sanos, aunque queda abierto por los canales 7 y 8  podr√≠an tener alguna relevancia y merecer un an√°lisis m√°s profundo.
# Conclusiones: 
1) No se encontr√≥ evidencia suficiente para afirmar que existen diferencias significativas en la energ√≠a promedio entre los sujetos del grupo de control y los sujetos con Parkinson en ninguno de los 8 canales analizados.
2) Aunque los canales 7 y 8 tienen valores p relativamente bajos, no llegan a ser significativos. Esto puede indicar que, aunque no se observa una diferencia clara en este an√°lisis, puede ser interesante hacer un an√°lisis m√°s profundo o probar con un mayor tama√±o de muestra.
3) Los datos de ambos grupos muestran una dispersi√≥n similar, lo que indica que la variabilidad es constante en ambos casos. Esto nos permite comparar directamente las medias de los grupos sin tener que preocuparnos por diferencias en la varianza.
