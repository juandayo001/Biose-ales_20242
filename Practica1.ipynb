```python
# PUNTOS J - k
n = np.arange(101)
y = np.sin(np.pi * 0.12 * n)
y2 = np.cos(2 * np.pi * 0.03 * n)
```

```python
# PUNTOS L
s = y + y2
t = y * y2
```

```python
plt.figure(figsize=(10, 6))
plt.plot(n, y, label='y[n] = sin(π*0.12n)', color='blue')
plt.plot(n, y2, label='y2[n] = cos(2π*0.03n)', color='red')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('y[n] and y2[n]')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(n, s, label='s[n] = y[n] + y2[n]', color='green')
plt.plot(n, t, label='t[n] = y[n] * y2[n]', color='purple')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('s[n] and t[n]')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
